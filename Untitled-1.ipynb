```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import xarray as xr
from glob import glob
```

```python
filename = '/gpfs/wolf2/cades/cli185/proj-shared/ywo/E3SM/output/20240311_US-SPR_ICB1850CNPRDCTCBC/run/lnd.log.34917.240830-174908'

file = open(filename)
```

```python
tvec = []
col_hummock = []
col_hollow = []

while True:
    line = file.readline()
    if not line:
        break

    # Remove trailing newline character
    line = line.strip()
    # Process the line here

    if line.startswith('Beginning timestep   : '):
        time = line.replace('Beginning timestep   : ', '')
        year = int(time[:4]) + 1800 # makes it possible for python to parse
        month = int(time[5:7])
        day = int(time[8:10])
        hour = int(time[11:13])

        if (year > 1900) & (year <= 2100):
            dt = pd.Timestamp(year = year, month = month, day = day, hour = hour)
            tvec.append(dt)

            line = file.readline().strip()
            while not (line.startswith('totvegp c')):
                line = file.readline().strip()

            line_data = np.array(re.split(r'\s+', line)[3:4]).astype(float)
            col_hummock.append(line_data)

            line = file.readline().strip()
            line_data = np.array(re.split(r'\s+', line)[3:4]).astype(float)
            col_hollow.append(line_data)
```

```python
file.close()
```

```python
tvec = pd.DatetimeIndex(tvec)
colnames = ['totvegp_col'] # ['totcolp', 'totpftp', 'cwdp', 'totlitp', 'totsomp', 'totprodp', 'solutionp', 'labilep', 'secondp', 'ptrunc', 'cropseedp_deficit']
col_hummock = pd.DataFrame(np.array(col_hummock), index = tvec, columns = colnames)
col_hollow = pd.DataFrame(np.array(col_hollow), index = tvec, columns = colnames)
```

```python
tvec = []
col_hummock_pima = []
col_hummock_lala = []
col_hummock_shrub = []
col_hummock_moss = []
col_hollow_pima = []
col_hollow_lala = []
col_hollow_shrub = []
col_hollow_moss = []

year = 0000

skip_beg = 0
while True:
    line = file.readline()
    if not line:
        break

    if skip_beg < 77964000:
        skip_beg = skip_beg + 1
        continue

    if line.startswith(' Beginning timestep   : '):
        time = line.replace(' Beginning timestep   : ', '')
        year = int(time[:4]) + 1800 # makes it possible for python to parse
        month = int(time[5:7])
        day = int(time[8:10])
        hour = int(time[11:13])
        dt = pd.Timestamp(year = year, month = month, day = day, hour = hour)

        if (year > 1900) & (year <= 2100):
            count = 0
            while count < 6:
                line = file.readline()

                if len(line) > 0 and 'fungi_som_to_ppool' in line:
                    line_data = re.split(r'\s+', line)[1:-1]

                    if count == 0:
                        col_hummock_pima.append(np.array(line_data[1:]).astype(float))
                    elif count == 1:
                        col_hummock_lala.append(np.array(line_data[1:]).astype(float))
                    elif count == 2:
                        col_hummock_shrub.append(np.array(line_data[1:]).astype(float))
                    elif count == 3:
                        col_hollow_pima.append(np.array(line_data[1:]).astype(float))
                    elif count == 4:
                        col_hollow_lala.append(np.array(line_data[1:]).astype(float))
                    elif count == 5:
                        col_hollow_shrub.append(np.array(line_data[1:]).astype(float))

                    count = count + 1
            tvec.append(dt)
```
